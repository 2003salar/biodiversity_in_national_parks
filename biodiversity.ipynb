```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
observations = pd.read_csv("observations.csv")
```

```python
species = pd.read_csv("species_info.csv")
# There are a lot of missing data in conservation_status
```

```python
misses = np.sum(species.conservation_status.isnull())
print((misses / len(species)) * 100)
```

```python
species.dropna(subset=["conservation_status"], inplace=True)
```

```python
species.reset_index(inplace=True, drop=True)
```

```python
# species.drop(columns=["level_0", "index"], inplace=True)
```

```python
species.groupby("conservation_status")["scientific_name"].count()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="conservation_status", data=species)
plt.xlabel("Conservation Status")
plt.ylabel("Number of Species")
plt.title("Conservation Status Count")
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="conservation_status", hue="category", data=species)
plt.xlabel("Conservation Status")
plt.ylabel("Number of Species")
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(10, 6))
filtered_data = observations.groupby("park_name")["observations"].sum().reset_index()
plt.bar(filtered_data["park_name"], filtered_data["observations"])
plt.xlabel("National Parks")
plt.ylabel("Total Observations")
plt.xticks(rotation=90)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
plt.clf()
```
